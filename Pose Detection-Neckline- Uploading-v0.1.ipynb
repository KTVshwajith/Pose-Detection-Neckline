```python
import cv2
import mediapipe as mp
import numpy as np
import math as m
from math import acos, degrees

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
mp_holistic = mp.solutions.holistic
pose = mp_pose.Pose()

cap = cv2.VideoCapture('Resources/lifting_02.mp4')
window_dect = 'Cap'
up = False
down = False
count = 0
with mp_pose.Pose(
     static_image_mode=False) as pose:

     while True:
        ret, frame = cap.read()
        if ret == False: 
            
            break
          #frame = cv2.flip(frame, 1)
        height, width, _ = frame.shape
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = pose.process(frame_rgb)
        
        lm = results.pose_landmarks
        lmPose = mp_pose.PoseLandmark

        if results.pose_landmarks is not None:
            
                        #---------------------Find the neckline---------------
            
            #left_shoulder define
            x31 = int(results.pose_landmarks.landmark[11].x * width)
            y31 = int(results.pose_landmarks.landmark[11].y * height)
            #Right_shoulder define
            x32 = int(results.pose_landmarks.landmark[12].x * width)
            y33 = int(results.pose_landmarks.landmark[12].y * height)
            #right eye outer define
            x33 = int(results.pose_landmarks.landmark[6].x * width)
            y33 = int(results.pose_landmarks.landmark[6].y * height)

            #x33 = int(results.pose_landmarks.landmark[32].x * width)
            #y33 = int(results.pose_landmarks.landmark[32].y * height)  
            
            l_shldr_x = int(lm.landmark[lmPose.LEFT_SHOULDER].x * width)
            l_shldr_y = int(lm.landmark[lmPose.LEFT_SHOULDER].y * height)
            # Right shoulder
            r_shldr_x = int(lm.landmark[lmPose.RIGHT_SHOULDER].x * width)
            r_shldr_y = int(lm.landmark[lmPose.RIGHT_SHOULDER].y * height)
            # Left ear.
            l_ear_x = int(lm.landmark[lmPose.LEFT_EAR].x * width)
            l_ear_y = int(lm.landmark[lmPose.LEFT_EAR].y * height)
            # Left hip.
            l_hip_x = int(lm.landmark[lmPose.LEFT_HIP].x * width)
            l_hip_y = int(lm.landmark[lmPose.LEFT_HIP].y * height)
            
            
            aux_image = np.zeros(frame.shape, np.uint8)
            #necklines
            cv2.line(aux_image, (l_shldr_x, l_shldr_y), (l_ear_x, l_ear_y), (255, 127, 0), 4)
            cv2.line(aux_image, (l_shldr_x, l_shldr_y), (l_shldr_x, l_shldr_y - 100), (255, 127, 0), 4)
            cv2.line(aux_image, (l_hip_x, l_hip_y), (l_shldr_x, l_shldr_y), (255, 127, 0), 4)
            cv2.line(aux_image, (l_hip_x, l_hip_y), (l_hip_x, l_hip_y - 100), (255, 127, 0), 4)
            
            
            output = cv2.addWeighted(frame, 1, aux_image, 0.8, 0)
            
            #Neckline            
            cv2.circle(output, (l_shldr_x, l_shldr_y), 6,(0, 255, 255), 4)
            cv2.circle(output, (l_ear_x, l_ear_y), 6, (0, 255, 255), 4)
            cv2.circle(output, (l_hip_x, l_hip_y), 6, (0, 255, 255), 4)
            
            
            def findDistance(x31, y31, x32, y32):
                dist = m.sqrt((x32-x31)**2+(y32-y31)**2)
                return dist  
            
            def findAngle(x32, y32, x33, y33):
                theta = m.acos((y33 -y32)*(-y32) / (m.sqrt((x33 - x32)**2 + (y33 - y32)**2) * y32))
                degree = int(180/m.pi)*theta
                return degree
            
            offset = findDistance(l_shldr_x, l_shldr_y, r_shldr_x, r_shldr_y)
            neck_inclination = findAngle(l_shldr_x, l_shldr_y, l_ear_x, l_ear_y)
            torso_inclination = findAngle(l_hip_x, l_hip_y, l_shldr_x, l_shldr_y)
            
            #cv2.putText(output, str(int(angle)), (x2 + 30, y2), 1, 1.5, (128, 255, 250), 2)
            
            #cv2.putText(output, angle_text_string, (10, 30), font, 0.9, (128, 255, 250), 2)
            cv2.putText(output, str(int(neck_inclination)), (l_shldr_x + 10, l_shldr_y), 1, 0.9, (128, 255, 250), 2)
            cv2.putText(output, str(int(torso_inclination)), (l_hip_x + 10, l_hip_y), 1, 0.9, (128, 255, 250), 2)
            
            
            fps = cap.get(5)
            frame_count = cap.get(7) 
            print('Frames per second : ', fps,'FPS')
            print('Frame count : ', frame_count)
            cv2.imshow('output', output)
            #cv2.imshow('Frame', frame)
            
            
       
        if cv2.waitKey(50) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
